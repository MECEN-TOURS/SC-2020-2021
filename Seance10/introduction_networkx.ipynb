# Construction  d'un graphe.

On va procéder par accumulation: commencer par instancier un graphe vide puis rajouter petit à petit des éléments.

## Syntaxe élémentaire

```python
mon_graphe = nx.Graph()
mon_graphe
```

```python
vars(mon_graphe)
```

```python
dir(mon_graphe)
```

**BILAN** les méthodes `add_node` et `add_edge` semblent toute désignées.

```python
help(mon_graphe.add_node)
```

```python
# Tout objet hashable (grosso modo immutable) doit permettre d'instancier un sommet (**node**)

mon_graphe.add_node("Premier Noeud")
mon_graphe.add_node(1)
mon_graphe.add_node((1, "a"))
vars(mon_graphe)
```

```python
help(mon_graphe.add_edge)
```

```python
# Rajouter une arrête rajoute automatiquement les sommets s'ils ne sont pas déjà présents

mon_graphe.add_edge("A", 1)
mon_graphe.add_edge("A", "B")
mon_graphe.add_edge((1, "a"), "B")
vars(mon_graphe)
```

## Gestion des attributs

On voit que les attributs internes  `_node` et `_adj` sont des dictionnaires, on peut fournir des attributs aux sommets et aux arrêtes

```python
# lors de la création, on verra plus tard comment les rajouter/modifier à postériori

mon_graphe.add_node("Nouveau", couleur="bleu", weight=15)
mon_graphe.add_edge(1, "B", temps=1, yolo=True)
vars(mon_graphe)
```

## Extension d'un graphe par bloc

On peut aussi utiliser les méthodes `add_nodes_from` et `add_edges_from` pour rajouter des sommets et arrêtes à partir d'itérateurs.

```python
help(mon_graphe.add_nodes_from)
```

```python
gr = nx.DiGraph()
gr.add_nodes_from([1, 2, 3])
gr.add_nodes_from("abc")
vars(gr)
```

```python
gr.add_edges_from(((1, 2), (1, 3), (1, "a"), ("a", "b"), ("a", "c")))
vars(gr)
```

**REMARQUE** on voit qu'un graphe dirigé contient deux dictionnaires de voisins
- un vers l'avant: `_succ`
- un vers l'arrière: `_pred`

# Sauvegarde et chargement

On peut choisir de lire/écrire les graphes dans des fichiers suivant différents format. 
Pour cela on utilisera soit
- le sous module `readwrite`
- les fonctions à la racine de `networkx` commençant par `read`/`write` (qui sont en fait importer du précédent module)

```python
help(nx.readwrite)
```

```python
original = nx.Graph()
```

```python
original.add_edges_from(
    [(i,j) for i in range(1, 11) for j in range(1, 11) if i >= j and i % j == 0]
)
original.adj
```

```python
nx.readwrite.adjlist.write_adjlist(original, path="./sauvegarde.txt")
```

```python
with open("./sauvegarde.txt", "r") as fichier:
    print(fichier.read())
```

```python
sauvegarde = nx.readwrite.adjlist.read_adjlist(path="./sauvegarde.txt")
```

```python
sauvegarde.adj
```

**ATTENTION** les différents formats n'ont pas les mêmes possibilités, en particulier certains ne permettent pas de stocker les éventuels attributs des sommets/arrêtes.
C'est le cas de `adjlist`.

```python
help(nx.readwrite.adjlist)
```

Un format complet mais un peu verbeux est  gexf (Graph Exchange XML Format), mais il y en beaucoup d'autres.

```python
autre = nx.Graph()
autre.add_node(1, label="Premier nombre entier?")
autre.add_node(10, label="Dernier considéré.")
autre.add_edge(1, 10, couleur="bleu")
autre.adj
```

```python
nx.readwrite.gexf.write_gexf(autre, path="test.gexf")
```

```python
with open("./test.gexf", "r") as fichier:
    print(fichier.read())
```

```python
recupere = nx.readwrite.gexf.read_gexf(path="test.gexf")
recupere.adj
```

**ATTENTION** le format `dot` étant spécifique à `graphviz` on y accède par `nx.drawing.nx_agraph`

```python
help(nx.drawing.nx_agraph)
```

# Visualisation d'un graphe

On utilise `nx.drawing` et de manière sous jacente `matplotlib`.

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = (12, 8)
```

```python
nombres = nx.Graph()
nombres.add_edges_from([
    (i, j) for i in range(2, 31) for j in range(2, 31) if i % j == 0
])
```

```python
print(len(nombres.nodes))
print(len(nombres.edges))
```

## Façon élémentaire

```python
nx.drawing.draw_networkx(G=nombres)
```

## Layout

On a en fait différentes fonctions permettant de calculer les positions où on placera les sommets.

```python
pos1 = nx.drawing.layout.spring_layout(nombres)
nx.drawing.draw_networkx(G=nombres, pos=pos1)
```

```python
pos1 = nx.drawing.layout.circular_layout(nombres)
nx.drawing.draw_networkx(G=nombres, pos=pos1)
```

```python
pos1 = nx.drawing.layout.kamada_kawai_layout(nombres)
nx.drawing.draw_networkx(G=nombres, pos=pos1)
```

```python
pos1 = nx.drawing.layout.random_layout(nombres)
nx.drawing.draw_networkx(G=nombres, pos=pos1)
```

```python
pos1 = nx.drawing.layout.spiral_layout(nombres)
nx.drawing.draw_networkx(G=nombres, pos=pos1)
```

## Gestion des poids

```python
oriente = nx.DiGraph()
a, b, c, d, e = "ABCDE"
for u, v, p in [(a, b, 5), (a, c, 4), (b, d, 2), (c, d, 1), (b, e, 1), (d, e, 3)]:
    oriente.add_edge(u, v, capacite=p)
```

```python
positions = nx.drawing.layout.shell_layout(G=oriente)
etiquettes = nx.get_edge_attributes(G=oriente, name="capacite")
nx.drawing.draw_networkx(G=oriente, pos=positions)
nx.drawing.draw_networkx_edge_labels(G=oriente, pos=positions, edge_labels=etiquettes);
```

# Premiers algorithmes

On a de nombreux algorithmes déjà implémentés. 
On prendra en exemple le graphe sur lequel on a expliquer Dijkstra.

<img src="../Seance08/graphe_base.svg">

```python
exemple = nx.Graph()
a,b,c,d,e,f,g,h,i = list("ABCDEFGHI")
exemple.add_edge(a, b, p=1)
exemple.add_edge(a, e, p=7)
exemple.add_edge(b, c, p=2)
exemple.add_edge(b, d, p=4)
exemple.add_edge(c, d, p=1)
exemple.add_edge(c, f, p=1)
exemple.add_edge(c, g, p=5)
exemple.add_edge(d, f, p=2)
exemple.add_edge(d, h, p=1)
exemple.add_edge(e, g, p=2)
exemple.add_edge(e, i, p=4)
exemple.add_edge(f, h, p=1)
exemple.add_edge(g, h, p=2)
exemple.add_edge(g, i, p=1)
exemple.adj
```

```python
pos = nx.drawing.layout.shell_layout(G=exemple)
nx.drawing.draw_networkx(G=exemple, pos=pos)
nx.drawing.draw_networkx_edge_labels(G=exemple, pos=pos);
```

```python
nx.algorithms.dijkstra_path_length(G=exemple, source="A", target="I", weight="p")
```

```python
nx.algorithms.dijkstra_path(G=exemple, source="A", target="I", weight="p")
```

```python
for resultat in nx.algorithms.all_pairs_dijkstra(exemple):
    print(resultat)
```
